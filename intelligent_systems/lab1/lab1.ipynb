## Населення

За даними на 2021 рік, населення Буркіна Фасо становить близько *21 мільйонів людей*. Більшість з них живуть у сільській місцевості, а найбільші міста - *Уагадугу* та *Бобо-Діуласо*.

## Державний устрій

Буркіна Фасо є республікою з *президентсько-парламентською* формою уряду. Президент є головним державним діячем та главою держави.

## Видатні персоналії

Однією з найвідоміших фігур в історії Буркіна Фасо є **Томас Санкара**, який був президентом з 1983 по 1987 рік. Він був відомий своїми спробами змінити політичну, економічну та соціальну структуру країни.

## Пам'ятки культури та природи

Буркіна Фасо має багату культурну спадщину, зокрема традиційне *мистецтво* та *музику*. Також у країні знаходиться багато національних парків та заповідників, де можна побачити різноманітні види тварин та рослин.

## Економіка

Буркіна Фасо є однією з *найбідніших* країн світу з низьким рівнем економічного розвитку. За даними Всесвітнього банку, у 2020 році *ВВП на душу населення склав $730*.

Однією з основних галузей економіки є сільське господарство, зокрема вирощування бавовни, зернових та фруктів. Також розвинені галузі добувної промисловості, зокрема з видобутку золота, марганцю та бокситів.

## Освіта

О*світа є однією з найбільших проблем* у Буркіна Фасо. Хоча освітня система є безкоштовною та обов'язковою для дітей віком від 6 до 16 років, проте доступ до неї обмежений, особливо у сільській місцевості. У 2019 році лише 22% населення мали більше, ніж початкову освіту.

## Здоров'я

Також здоров'я є однією з *проблем* у Буркіна Фасо. У країні досить високі рівні смертності від хвороб, пов'язаних з брудною водою та нездоровими умовами життя. Також у країні є проблема зі зниженням рівня вакцинації на деякі вірусні хвороби, такі як поліомієліт та дифтерія.

## Культура

Культура Буркіна Фасо є дуже *різноманітною*, зокрема в *традиційному мистецтві*, яке відображається у *музиці*, *танцях* та *скульптурах*. Один з видатних свят у країні - це **Фестиваль масок у Дедуґу**, який відбувається кожні два роки та відзначає традиційне мистец

```python
def gcd_sub(a, b):
    while b != 0:
        if a > b:
            a = a - b
        else:
            b = b - a

    return a


def gcd_div(a, b):
    while b != 0:
        a, b = b, a % b

    return a
```
